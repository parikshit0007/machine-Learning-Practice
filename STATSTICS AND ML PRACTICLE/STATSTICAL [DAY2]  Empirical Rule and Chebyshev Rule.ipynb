# Empirical Rule :

```python

# mean  & Standerd Deviation:

mean = 0
SD = 1
```

```python
# value of CDF [' Cummilitive distrubution Function']
# between one, two, and three S.D around the mean
```

```python
one_sd = norm.cdf(SD,mean,SD) - norm.cdf(-SD,mean,SD)
two_sd = norm.cdf(2*SD,mean,SD)- norm.cdf(-2*SD,mean,SD)
three_sd = norm.cdf(2*SD,mean,SD)- norm.cdf(-3*SD,mean,SD)

print("The First Standerd Deviation is : ", one_sd)
print("The Second Standerd Deviation is : ", two_sd)
print("The Three Standerd Deviation is : ", three_sd)
```
